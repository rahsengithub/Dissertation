```python
import numpy as np
hour = np.arange(0,100,1.0) %24
print(hour)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
val  = np.arange(0, len(hour), 1.0)
ran = np.random.randn(len(hour))
new = 0.1*(val + ran)
cosval = np.cos(2*np.pi*hour/24)
cosval12 = np.cos(2*np.pi*hour/12)
sinval = np.sin(2*np.pi*hour/24)
sinval12 = np.sin(2*np.pi*hour/12)
X = np.c_[cosval, sinval, cosval12, sinval12, new]
# plt.plot(X)
X = X / np.std(X, 0)
# plt.plot(X)
X = X - np.mean(X,0)
# plt.plot(X)
X = np.c_[X,np.ones(len(X))]
plt.figure(figsize=[20,20])
plt.imshow(X)
```

```python
y = np.zeros(len(X))

y[hour<6] = 4

print(y)
print(new)
y = y+new*0.5 + np.random.rand(len(y))
print(y)
```

```python
plt.plot(y)
```

```python
w = np.linalg.inv(X.T @ X) @ X.T @ y
plt.plot(y,'r-')
plt.plot(X@w)
print(w)
np.mean(np.abs(X@w-y)) #mean abs error
```
